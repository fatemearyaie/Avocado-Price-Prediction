## Specifying dependent and independent variables

```python
X = df.drop(['AveragePrice'], axis = 1)
y = df['AveragePrice']
```

### Labeling the categorical variables

```python
Xcat = pd.get_dummies(X[["type","region"]], drop_first = True)
```

```python
Xnum = X[["Small Hass","Large Hass","XLarge Hass","Small Bags","Large Bags","XLarge Bags"]]
```

```python
X = pd.concat([Xcat, Xnum], axis = 1)
```

```python
X.shape
```

```python
F_DF = pd.concat([y,X],axis=1)
F_DF.head(2)
```

### visualise the highly correlated Variables with the Average Prices

```python
sns.set(color_codes=True)
sns.jointplot(x="Small Hass", y="AveragePrice", data=F_DF, kind="reg")
```

```python
sns.jointplot(x="Small Bags", y="AveragePrice", data=F_DF, kind="reg")
```

```python
sns.jointplot(x="Large Bags", y="AveragePrice", data=F_DF, kind="reg")
```

```python
sns.lmplot(x="Small Hass", y="AveragePrice", col="type_organic", data=F_DF, col_wrap=2)
```

## Train And Test data

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train.shape
```

## Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

```python
lm_predict = lm.predict(X_test)
```

## Evaluation

```python
print ("R2 of Linear Regresson:", lm.score(X_train,Y_train))
```

```python
print('MAE: ',metrics.mean_absolute_error(Y_test, lm_predict))
```

```python
print('MSE: ',metrics.mean_squared_error(Y_test,lm_predict))
```

```python
print('RMSE:',np.sqrt(metrics.mean_squared_error(Y_test, lm_predict)))
```

```python
plt.figure(figsize=(6,4))
sns.distplot(Y_test - lm_predict)
plt.title('Distribution of residuals');
```
